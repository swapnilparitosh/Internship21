```python
import requests
from bs4 import BeautifulSoup
```

```python
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
header = soup.find_all(['h1', 'h2','h3','h4','h5','h6'])
for i in header:
    print(i.text)
```
