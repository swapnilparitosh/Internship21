```python
import requests
from bs4 import BeautifulSoup
```

```python
page=requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
data =soup.findAll('div', class_="restnt-card restaurant")
```

```python
restaurant_name=[]
cuisine=[]
location=[]
rating=[]
url=[]
```

```python
for i in data:
    restaurant_name.append(i.find('div', class_="restnt-info cursor").text.split(',')[0])
    
    cuisine.append(i.find('span', class_="double-line-ellipsis").text.split('|')[1])

    location.append(i.find('div', class_="restnt-info cursor").text.split(',')[1])
    
    rating.append(i.find('div', class_='restnt-rating rating-3'))
    
    url.append(i.find('img', class_="no-img")['data-src'])
    
print(restaurant_name,'\n\n\n')
print(cuisine,'\n\n\n')
print(location,'\n\n\n')
print(rating,'\n\n\n')
print(url,'\n\n\n')
```

```python
import pandas as pd
df=pd.DataFrame({'Restaurant Name':restaurant_name, 'Cuisine':cuisine, 'Location':location, 'rating':rating, 'Image URL': url})
df
```
